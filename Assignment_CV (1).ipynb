```python
import os
import numpy as np
import cv2
import numpy as np 
from skimage.feature import hog 

features = [] 
labels = []


directory = "qw"
dimension = (64, 64)

#Reading files from directory
for folder in os.listdir(directory):
    print("Processing Folder = ",folder)
    name1= directory + folder + "/"
    print(name1)
    for file in os.listdir(directory + "/" +folder + "/"):

        name = directory + "/" + folder +"/" + file
        
        #Reading image using OpenCV
        image = cv2.imread(name)
       
        if image is not None:
            print(name)
            #Reszing image so that all image dimensions are consistent
            img = cv2.resize(image, dimension, interpolation = cv2.INTER_AREA)

            #Using HOG to extract images from resized images
            derived_features, _ = hog(img, orientations=9, pixels_per_cell=(8, 8), 
                        cells_per_block=(2, 2), visualize=True, multichannel=True)

            features.append(derived_features) 

            labels.append(int(folder))

        else:
            print("Error Opening image")
        pass
```

# Training Random Forest Classifier using features from HOG

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Converting features into Numpy array
features = np.array(features) 
#Using Random Forest Classifier with 20 trees
lin_clf = RandomForestClassifier(n_estimators = 20) 
#Training the classifier using new derived features and training labels 
lin_clf.fit(features, labels)
```

# Testing

```python
directory = "qw"
dimension = (64, 64)

tp = 0 # True Positive
fp = 0 # False Positive
fn = 0 # False Negative


for folder in os.listdir(directory):
    print("Processing Folder = ",folder)
    
    for file in os.listdir(directory + "/" + folder + "/"):

        name = directory + "/" + folder + "/" + file
        #Reading images using opencv
        image = cv2.imread(name)
       
        if image is not None:
            # Resizing the image so that dimensions of images are consistent
            img = cv2.resize(image, dimension, interpolation = cv2.INTER_AREA)

            fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), 
                        cells_per_block=(2, 2), visualize=True, multichannel=True)

            prediction = lin_clf.predict([fd])
            print("Pred = ",prediction[0])
            print("Label = ", int(folder))
            if prediction[0] == int(folder):
                tp += 1
            else:
                fp += 1
                fn += 1

        else:
            print("Error Opening image")
        pass
```

# Performance Measures

## Precision

```python
precision = tp/(tp+fp)
print("Precision = ", precision)
```

```python
recall = tp/(tp + fn)
print("Recall = ", recall)
```

```python
f_1 = 2*(precision * recall)/(precision + recall)
print("F_1 Score = ", f_1)
```
